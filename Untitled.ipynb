```python
import pandas as pd
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
#Drop unnessary column
df.drop('CustomerID', axis = 1, inplace=True)
```

```python
#Let select importent feature for clustering
df1 = df[['Annual Income (k$)',	'Spending Score (1-100)']]
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'])
```

```python
#let scaling the columns
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df1[['Annual Income (k$)']])
df1['Annual Income (k$)'] = scaler.fit_transform(df1[['Annual Income (k$)']])
#
```

```python
scaler.fit(df1[['Spending Score (1-100)']])
df1['Spending Score (1-100)'] = scaler.fit_transform(df1[['Spending Score (1-100)']])
```

```python
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'])
```

```python
from sklearn.cluster import KMeans
```

```python
#find the k value
sse = []
#k_range = range(1,11)

for k in range(1, 11):
    km = KMeans(n_clusters = k)
    km.fit(df1[['Annual Income (k$)','Spending Score (1-100)']])
    sse.append(km.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1, 11),sse, color= 'green', marker='*')
plt.xlabel('No of Clusters')
plt.ylabel('sse')
plt.title('Elbow Method')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

from above Elbow plot we can say that the optimal size of the cluster is that is k=5

```python
km = KMeans(n_clusters=5)
y_predicted = km.fit_predict(df1[['Annual Income (k$)','Spending Score (1-100)']])
print(y_predicted)
df1['cluster'] = y_predicted
df1.head()
```

```python
df1['cluster'].unique()
```

```python
km.cluster_centers_
```

```python
df2 = df1[df1['cluster']==0]
df3 = df1[df1['cluster']==1]
df4 = df1[df1['cluster']==2]
df5 = df1[df1['cluster']==3]
df6 = df1[df1['cluster']==4]
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color = 'red', label = 'clusler0')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color = 'green', label = 'clusler1')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color = 'black', label = 'clusler2')
plt.scatter(df5['Annual Income (k$)'], df5['Spending Score (1-100)'], color = 'blue', label = 'clusler3')
plt.scatter(df6['Annual Income (k$)'], df6['Spending Score (1-100)'], color = 'y', label = 'clusler4')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title("cluster of customer")
plt.legend()
```
